### Separate categorical and numeric columns

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
    """
    grab_col_names for given dataframe

    :param dataframe:
    :param cat_th:
    :param car_th:
    :return:
    """

    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]

    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]

    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]

    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')

    

    return cat_cols, cat_but_car, num_cols, num_but_cat

cat_cols, cat_but_car, num_cols, num_but_cat = grab_col_names(df)
```

## Outlier Analysis for Numerical Columns

```python
num_cols
```

```python
# gereksiz sütunları sil
num_cols = [col for col in num_cols if col not in ['Unnamed: 0','id']]
num_cols
```

```python
def outlier_threshold(dataframe,col_name,q1=0.10,q3=0.90):
    Q1 = dataframe[col_name].quantile(q1)
    Q3 = dataframe[col_name].quantile(q3)

    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    return lower_limit,upper_limit
    
```

```python
def check_outlier(dataframe,col_name):
    lower_limit,upper_limit =  outlier_threshold(dataframe,col_name)
    if dataframe[(dataframe[col_name] < lower_limit)| (dataframe[col_name] > upper_limit)].any(axis=None):
        return True
    else:
        return False
```

```python
for col in num_cols:
    print(col,check_outlier(df,col))
    
```

```python

# aykırı değerlerler threshold value kullanılarak limit değerlere eşitlenir
def replace_with_thresholds(dataframe,col_name):
    dataframe[col_name] = dataframe[col_name].astype(float)
    low_limit, up_limit = outlier_threshold(dataframe,col_name)
    dataframe.loc[(dataframe[col_name] < low_limit),col_name] = low_limit
    dataframe.loc[(dataframe[col_name] > up_limit),col_name] = up_limit
```

```python
for col in num_cols:
    replace_with_thresholds(df,col)
```

```python
for col in num_cols:
    print(col,check_outlier(df,col))
# Başarılı 
```


## Veri Temizleme ve Hazırlık (Data Cleaning & Preparation)
* Arrival Delay in Minutes sütununda eksik veriler bulunmuştur.Bu eksik veriler ortalama alma yöntemiyle ( df['Arrival Delay in Minutes'].mean() ) doldurulmuştur.
* Sayısal sütunlar için Aykırı Değer analizi yapılmıştır
* Veri setindeki sayısal değişkenlerde uç değerleri belirlemek için IQR (Interquartile Range / Çeyrekler Arası Aralık) yöntemini kullanıldı. Bu yöntemle:
1) Q1 (1. çeyrek) ve Q3 (3. çeyrek) değerlerini hesaplandı.Default olarak Q1= 0.25 ,Q3 = 0.75 kullanıldı
2) IQR = Q3 - Q1 formülüyle çeyrekler arası aralığı bulunuldu.
3) Uç değer sınırlarını aşağıdaki şekilde tanımlandı:
 * Alt sınır = Q1 - 1.5 × IQR
 * Üst sınır = Q3 + 1.5 × IQR
* Bu sınırların altında veya üstünde kalan değerleri uç değer (outlier) olarak kabul edildi.
* Gözlemlenen bazı değişkenlerde (örneğin Departure Delay in Minutes ve Arrival Delay in Minutes) ciddi sayıda uç değer vardı.
Uç değerleri tamamen silmek yerine, analizde sapmalara neden olmamaları için bu değerler belirlenen alt ve üst sınırlarla sınırlandı (capping).
Böylece hem veri kaybı yaşanmadı hem de uç değerlerin olumsuz etkisi azaltılmış oldu.
